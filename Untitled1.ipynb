# Application de gestion des inscriptions des étudiants du centre universitaire de koula-moutou

```python
# ==============================================================================
# MODULE D'IMPORTATION DES BIBLIOTHÈQUES
# ==============================================================================

# Importation de tkinter pour créer l'interface graphique (fenêtres, boutons, etc.)
import tkinter as tk

# Importation de messagebox pour afficher des boîtes de dialogue (erreurs, succès, etc.)
from tkinter import messagebox

# Importation de datetime pour valider et formater les dates de naissance
from datetime import datetime


# ==============================================================================
# VARIABLE GLOBALE - LISTE DES ÉTUDIANTS
# ==============================================================================

# Cette liste vide va contenir tous les étudiants ajoutés
# Chaque étudiant sera un dictionnaire avec toutes ses informations
etudiants = []


# ==============================================================================
# CLASSE PRINCIPALE DE L'APPLICATION
# ==============================================================================

# Définition de la classe Application qui hérite de tk.Tk (fenêtre principale)
class Application(tk.Tk):
    
    # ------------------------------------------------------------------------------
    # Constructeur de la classe - s'exécute quand on crée l'application
    # ------------------------------------------------------------------------------
    def __init__(self):
        # Appel du constructeur de la classe parente (tk.Tk)
        super().__init__()
        
        # Définition du titre de la fenêtre
        self.title("Gestion des Inscriptions - Université")
        
        # Définition de la taille de la fenêtre (largeur x hauteur en pixels)
        self.geometry("600x500")
        
        # Définition de la couleur de fond de la fenêtre (gris clair)
        self.configure(bg="#f0f0f0")
        
        # Création d'un cadre (frame) pour contenir tous les widgets de l'interface
        self.frame = tk.Frame(self, bg="#f0f0f0")
        
        # Placement du cadre dans la fenêtre avec des marges (padx=20, pady=20)
        # expand=True permet au cadre de s'étendre, fill="both" remplit l'espace
        self.frame.pack(expand=True, fill="both", padx=20, pady=20)
        
        # Appel de la méthode pour afficher le menu principal
        self.afficher_menu()
    
    
    # ------------------------------------------------------------------------------
    # Méthode afficher_menu - Affiche le menu principal avec les 5 choix
    # ------------------------------------------------------------------------------
    def afficher_menu(self):
        # Effacement de tous les widgets du cadre pour préparer l'affichage du menu
        self.clear_frame()
        
        # Création d'un label (texte) pour le titre du menu
        # font=("Arial", 18, "bold") définit la police, la taille et le gras
        # fg="#2c3e50" définit la couleur du texte (bleu foncé)
        tk.Label(self.frame, text="GESTION DES INSCRIPTIONS", font=("Arial", 18, "bold"), bg="#f0f0f0", fg="#2c3e50").pack(pady=20)
        
        # Définition de la liste des boutons du menu
        # Chaque élément est un tuple (texte_du_bouton, fonction_à_appeler)
        boutons = [
            ("1 - Ajouter un étudiant", self.ajouter_etudiant),
            ("2 - Afficher les étudiants", self.afficher_etudiants),
            ("3 - Modifier un étudiant", self.modifier_etudiant),
            ("4 - Supprimer un étudiant", self.supprimer_etudiant),
            ("5 - Quitter", self.quitter)
        ]
        
        # Boucle pour créer chaque bouton du menu
        for texte, commande in boutons:
            # Création d'un bouton avec le texte et la commande associés
            # width=25 définit la largeur, height=2 définit la hauteur
            # bg et fg définissent les couleurs de fond et de texte
            # relief="raised" donne un effet de surélévation au bouton
            btn = tk.Button(self.frame, text=texte, font=("Arial", 12), width=25, height=2, 
                          command=commande, bg="#3498db", fg="white", relief="raised")
            
            # Placement du bouton avec une marge verticale (pady=5)
            btn.pack(pady=5)
    
    
    # ------------------------------------------------------------------------------
    # Méthode clear_frame - Efface tous les widgets du cadre
    # ------------------------------------------------------------------------------
    def clear_frame(self):
        # Récupération de tous les widgets enfants du cadre
        # winfo_children() retourne une liste de tous les widgets contenus
        for widget in self.frame.winfo_children():
            # Destruction de chaque widget pour effacer l'écran
            widget.destroy()
    
    
    # ------------------------------------------------------------------------------
    # Méthode retour_menu - Affiche un bouton pour revenir au menu principal
    # ------------------------------------------------------------------------------
    def retour_menu(self):
        # Création d'un bouton "Retour au menu" qui appelle afficher_menu
        tk.Button(self.frame, text="Retour au menu", command=self.afficher_menu, bg="#95a5a6", fg="white").pack(pady=20)
    
    
    # ------------------------------------------------------------------------------
    # Méthode ajouter_etudiant - Affiche le formulaire d'ajout d'un étudiant
    # ------------------------------------------------------------------------------
    def ajouter_etudiant(self):
        # Effacement du cadre pour afficher le formulaire
        self.clear_frame()
        
        # Affichage du titre du formulaire
        tk.Label(self.frame, text="AJOUTER UN ÉTUDIANT", font=("Arial", 16, "bold"), bg="#f0f0f0", fg="#2c3e50").pack(pady=10)
        
        # Définition de la liste des labels (titres des champs à remplir)
        labels = ["Nom", "Prénom", "Sexe (M/F)", "Date de naissance (JJ/MM/AAAA)", 
                 "Matricule", "Filière", "Niveau", "Nom du père", "Nom de la mère"]
        
        #Initialisation d'un dictionnaire pour stocker les champs de saisie (Entry)
        self.entries = {}
        
        # Boucle pour créer chaque champ de saisie
        for label in labels:
            # Affichage du label (titre du champ)
            tk.Label(self.frame, text=label, bg="#f0f0f0").pack(anchor="w")
            
            # Création d'un champ de saisie (Entry) de largeur 40 caractères
            entry = tk.Entry(self.frame, width=40)
            
            # Placement du champ de saisie
            entry.pack(pady=2)
            
            # Ajout du champ dans le dictionnaire avec le label comme clé
            self.entries[label] = entry
        
        # Création d'un cadre pour contenir les boutons
        btn_frame = tk.Frame(self.frame, bg="#f0f0f0")
        
        # Placement du cadre des boutons
        btn_frame.pack(pady=15)
        
        # Création du bouton "Ajouter" qui appelle valider_ajout
        # bg="#27ae60" définit la couleur verte pour indiquer une action positive
        tk.Button(btn_frame, text="Ajouter", command=self.valider_ajout, bg="#27ae60", fg="white").pack(side="left", padx=5)
        
        # Ajout du bouton pour revenir au menu
        self.retour_menu()
    
    
    # ------------------------------------------------------------------------------
    # Méthode valider_ajout - Valide et ajoute un étudiant dans la liste
    # ------------------------------------------------------------------------------
    def valider_ajout(self):
        # Utilisation de try/except pour capturer les erreurs de validation
        try:
            # Initialisation d'un dictionnaire pour stocker les données saisies
            donnees = {}
            
            # Boucle pour récupérer toutes les valeurs saisies dans les champs
            # .get() récupère le texte entré, .strip() enlève les espaces au début et à la fin
            for label, entry in self.entries.items():
                donnees[label] = entry.get().strip()
            
            # Boucle pour vérifier que tous les champs sont remplis
            for key, value in donnees.items():
                # Si un champ est vide
                if not value:
                    # Affichage d'un message d'erreur avec le nom du champ vide
                    messagebox.showerror("Erreur", f"Le champ {key} est obligatoire")
                    
                    # Arrêt de la fonction
                    return
            
            # Validation du format de la date de naissance
            # datetime.strptime analyse la chaîne selon le format "%d/%m/%Y"
            # Ce format signifie: jour/mois/année avec 2 chiffres chacun
            date_naiss = datetime.strptime(donnees["Date de naissance (JJ/MM/AAAA)"], "%d/%m/%Y")
            
            # Création du dictionnaire représentant l'étudiant
            # On y met toutes les informations récupérées
            etudiant = {
                "nom": donnees["Nom"],                       # Nom de l'étudiant
                "prenom": donnees["Prénom"],                 # Prénom de l'étudiant
                "sexe": donnees["Sexe (M/F)"],               # Sexe (M ou F)
                "date_naissance": date_naiss.strftime("%d/%m/%Y"),  # Date au format JJ/MM/AAAA
                "matricule": donnees["Matricule"],            # Numéro de matricule unique
                "filiere": donnees["Filière"],              # Filière d'études
                "niveau": donnees["Niveau"],                # Niveau d'études
                "nom_pere": donnees["Nom du père"],          # Nom du père
                "nom_mere": donnees["Nom de la mère"]         # Nom de la mère
            }
            
            # Ajout de l'étudiant dans la liste globale des étudiants
            # .append() ajoute l'élément à la fin de la liste
            etudiants.append(etudiant)
            
            # Affichage d'un message de succès
            messagebox.showinfo("Succès", "Étudiant ajouté avec succès")
            
            # Retour au menu principal
            self.afficher_menu()
            
        # Bloc except qui s'exécute si une erreur se produit (ici: format de date invalide)
        except ValueError:
            # Affichage d'un message d'erreur
            messagebox.showerror("Erreur", "Format de date invalide. Utiliser JJ/MM/AAAA")
    
    
    # ------------------------------------------------------------------------------
    # Méthode afficher_etudiants - Affiche la liste de tous les étudiants
    # ------------------------------------------------------------------------------
    def afficher_etudiants(self):
        # Effacement du cadre
        self.clear_frame()
        
        # Affichage du titre
        tk.Label(self.frame, text="LISTE DES ÉTUDIANTS", font=("Arial", 16, "bold"), bg="#f0f0f0", fg="#2c3e50").pack(pady=10)
        
        # Vérification si la liste des étudiants est vide
        if not etudiants:
            # Si la liste est vide, affichage d'un message d'information
            tk.Label(self.frame, text="Aucun étudiant enregistré", bg="#f0f0f0", fg="#e74c3c").pack()
        else:
            # ------------------------------------------------------------------------------
            # SI LA LISTE N'EST PAS VIDE - Affichage avec scrollbar (défilement)
            # ------------------------------------------------------------------------------
            
            # Création d'un_canvas (zone de dessin) pour le défilement
            canvas = tk.Canvas(self.frame, bg="#f0f0f0")
            
            # Création d'une scrollbar (barre de défilement) verticale
            # command=canvas.yview permet de défilier verticalement
            scrollbar = tk.Scrollbar(self.frame, orient="vertical", command=canvas.yview)
            
            # Création d'un cadre secondaire qui contendra les étudiants
            scrollable_frame = tk.Frame(canvas, bg="#f0f0f0")
            
            # Liaison de l'événement de configuration du cadre au calcul de la zone de défilement
            # Cela met à jour la zone de défilement quand le contenu change
            scrollable_frame.bind("<Configure>", lambda e: canvas.configure(scrollregion=canvas.bbox("all")))
            
            # Création d'une fenêtre interne dans le canvas à la position (0,0)
            # anchor="nw" signifie "north-west" (coin supérieur gauche)
            canvas.create_window((0, 0), window=scrollable_frame, anchor="nw")
            
            # Configuration de la scrollbar pour contrôler le canvas
            canvas.configure(yscrollcommand=scrollbar.set)
            
            # Placement du canvas à gauche, avec remplissage total
            canvas.pack(side="left", fill="both", expand=True, pady=10)
            
            # Placement de la scrollbar à droite, avec remplissage vertical
            scrollbar.pack(side="right", fill="y")
            
            # Boucle pour afficher chaque étudiant
            for i, et in enumerate(etudiants):
                # Création de la chaîne de texte avec toutes les infos de l'étudiant
                info = (f"Nom: {et['nom']} | Prénom: {et['prenom']} | "
                       f"Date de naissance: {et['date_naissance']} | "
                       f"Matricule: {et['matricule']} | "
                       f"Filière: {et['filiere']} | Niveau: {et['niveau']}")
                
                # Création d'un label pour afficher les informations de l'étudiant
                # width=70 définit la largeur, anchor="w"	aligne à gauche
                # bg et fg définissent les couleurs
                tk.Label(scrollable_frame, text=info, bg="#ecf0f1", fg="#2c3e50", 
                        width=70, anchor="w", padx=5, pady=5).pack(fill="x", pady=2)
        
        # Ajout du bouton de retour au menu
        self.retour_menu()
    
    
    # ------------------------------------------------------------------------------
    # Méthode modifier_etudiant - Affiche le formulaire de recherche pour modification
    # ------------------------------------------------------------------------------
    def modifier_etudiant(self):
        # Effacement du cadre
        self.clear_frame()
        
        # Affichage du titre
        tk.Label(self.frame, text="MODIFIER UN ÉTUDIANT", font=("Arial", 16, "bold"), bg="#f0f0f0", fg="#2c3e50").pack(pady=10)
        
        # Affichage du label pour le champ de recherche
        tk.Label(self.frame, text="Matricule de l'étudiant à modifier:", bg="#f0f0f0").pack(anchor="w")
        
        # Création du champ de saisie pour le matricule
        self.entry_matricule = tk.Entry(self.frame, width=40)
        
        # Placement du champ de saisie
        self.entry_matricule.pack(pady=5)
        
        # Création du bouton "Rechercher" qui appelle rechercher_modifier
        # bg="#f39c12" définit la couleur orange
        tk.Button(self.frame, text="Rechercher", command=self.rechercher_modifier, bg="#f39c12", fg="white").pack(pady=10)
        
        # Ajout du bouton de retour au menu
        self.retour_menu()
    
    
    # ------------------------------------------------------------------------------
    # Méthode rechercher_modifier - Cherche l'étudiant et affiche le formulaire de modification
    # ------------------------------------------------------------------------------
    def rechercher_modifier(self):
        # Récupération du matricule entré par l'utilisateur
        matricule = self.entry_matricule.get().strip()
        
        # Boucle pour rechercher l'étudiant dans la liste
        for i, et in enumerate(etudiants):
            # Si le matricule correspond
            if et["matricule"] == matricule:
                # Effacement du cadre pour afficher le formulaire de modification
                self.clear_frame()
                
                # Affichage du titre avec le nom de l'étudiant à modifier
                tk.Label(self.frame, text=f"Modifier: {et['nom']} {et['prenom']}", 
                        font=("Arial", 14), bg="#f0f0f0", fg="#2c3e50").pack(pady=10)
                
                # Affichage du label pour la nouvelle filière
                tk.Label(self.frame, text="Nouvelle filière:", bg="#f0f0f0").pack(anchor="w")
                
                # Création du champ de saisie pour la nouvelle filière
                self.entry_filiere = tk.Entry(self.frame, width=40)
                
                # Insertion de la filière actuelle dans le champ (valeur par défaut)
                self.entry_filiere.insert(0, et["filiere"])
                
                # Placement du champ de saisie
                self.entry_filiere.pack(pady=5)
                
                # Affichage du label pour le nouveau niveau
                tk.Label(self.frame, text="Nouveau niveau:", bg="#f0f0f0").pack(anchor="w")
                
                # Création du champ de saisie pour le nouveau niveau
                self.entry_niveau = tk.Entry(self.frame, width=40)
                
                # Insertion du niveau actuel dans le champ (valeur par défaut)
                self.entry_niveau.insert(0, et["niveau"])
                
                # Placement du champ de saisie
                self.entry_niveau.pack(pady=5)
                
                # Stockage de l'index de l'étudiant à modifier dans une variable d'instance
                self.index_modifier = i
                
                # Création du bouton pour valider les modifications
                tk.Button(self.frame, text="Enregistrer les modifications", 
                         command=self.valider_modification, bg="#27ae60", fg="white").pack(pady=15)
                
                # Ajout du bouton de retour au menu
                self.retour_menu()
                
                # Arrêt de la fonction après avoir trouvé l'étudiant
                return
        
        # Si aucune correspondance n'a été trouvée, affichage d'un message d'erreur
        messagebox.showerror("Erreur", "Étudiant introuvable")
    
    
    # ------------------------------------------------------------------------------
    # Méthode valider_modification - Enregistre les modifications apportées à un étudiant
    # ------------------------------------------------------------------------------
    def valider_modification(self):
        # Récupération de la nouvelle fili��re saisie
        nouvelle_filiere = self.entry_filiere.get().strip()
        
        # Récupération du nouveau niveau saisi
        nouveau_niveau = self.entry_niveau.get().strip()
        
        # Vérification que les deux champs sont remplis
        if not nouvelle_filiere or not nouveau_niveau:
            # Si un champ est vide, affichage d'un message d'erreur
            messagebox.showerror("Erreur", "Tous les champs sont obligatoires")
            
            # Arrêt de la fonction
            return
        
        # Modification de la filière de l'étudiant dans la liste
        # On accède à l'étudiant par son index et on modifie la clé "filiere"
        etudiants[self.index_modifier]["filiere"] = nouvelle_filiere
        
        # Modification du niveau de l'étudiant dans la liste
        etudiants[self.index_modifier]["niveau"] = nouveau_niveau
        
        # Affichage d'un message de succès
        messagebox.showinfo("Succès", "Modifications effectuées")
        
        # Retour au menu principal
        self.afficher_menu()
    
    
    # ------------------------------------------------------------------------------
    # Méthode supprimer_etudiant - Affiche le formulaire de suppression
    # ------------------------------------------------------------------------------
    def supprimer_etudiant(self):
        # Effacement du cadre
        self.clear_frame()
        
        # Affichage du titre
        tk.Label(self.frame, text="SUPPRIMER UN ÉTUDIANT", font=("Arial", 16, "bold"), bg="#f0f0f0", fg="#2c3e50").pack(pady=10)
        
        # Affichage du label pour le champ de recherche
        tk.Label(self.frame, text="Matricule de l'étudiant à supprimer:", bg="#f0f0f0").pack(anchor="w")
        
        # Création du champ de saisie pour le matricule
        self.entry_matricule_suppr = tk.Entry(self.frame, width=40)
        
        # Placement du champ de saisie
        self.entry_matricule_suppr.pack(pady=5)
        
        # Création du bouton "Supprimer" qui appelle valider_suppression
        # bg="#e74c3c" définit la couleur rouge pour indiquer une action destructive
        tk.Button(self.frame, text="Supprimer", command=self.valider_suppression, bg="#e74c3c", fg="white").pack(pady=10)
        
        # Ajout du bouton de retour au menu
        self.retour_menu()
    
    
    # ------------------------------------------------------------------------------
    # Méthode valider_suppression - Supprime l'étudiant de la liste
    # ------------------------------------------------------------------------------
    def valider_suppression(self):
        # Récupération du matricule entré par l'utilisateur
        matricule = self.entry_matricule_suppr.get().strip()
        
        # Boucle pour recherche l'étudiant dans la liste
        for i, et in enumerate(etudiants):
            # Si le matricule correspond
            if et["matricule"] == matricule:
                # Suppression de l'étudiant de la liste par son index
                # .pop(i) retire l'élément à l'index i
                etudiants.pop(i)
                
                # Affichage d'un message de succès
                messagebox.showinfo("Succès", "Étudiant supprimé avec succès")
                
                # Retour au menu principal
                self.afficher_menu()
                
                # Arrêt de la fonction après la suppression
                return
        
        # Si aucune correspondance n'a été trouvée, affichage d'un message d'erreur
        messagebox.showerror("Erreur", "Étudiant non trouvé")
    
    
    # ------------------------------------------------------------------------------
    # Méthode quitter - Ferme l'application
    # ------------------------------------------------------------------------------
    def quitter(self):
        # Affichage d'une boîte de dialogue de confirmation
        # askokcancel retourne True si l'utilisateur clique "OK"
        if messagebox.askokcancel("Quitter", "Voulez-vous vraiment quitter?"):
            # Fermeture de la fenêtre (arrêt de l'application)
            self.destroy()


# ==============================================================================
# POINT D'ENTRÉE DU PROGRAMME
# ==============================================================================

# Cette condition vérifie si le fichier est exécuté directement (pas importé)
if __name__ == "__main__":
    # Création d'une instance de l'application
    app = Application()
    
    # Lancement de la boucle principale de l'application
    # Cette boucle gère les événements (clics, saisies, etc.) jusqu'à la fermeture
    app.mainloop()
```
